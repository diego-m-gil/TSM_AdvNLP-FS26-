**Topic:** Classification performance evaluation (Lecture 4) — Swiss-cities style

**Walkthrough:** Two examples with GT vs classifier columns -> confusion matrix -> per-class P/R/F1 -> macro-F1 and weighted-F1 -> plots.

**Remember for exam:** P=TP/(TP+FP), R=TP/(TP+FN), F1=2PR/(P+R). N/A when 0/0. macro=avg of F1s; weighted=avg by support. THIS IS THE EXACT EXAM CALC STYLE.

**Time:** 30min practice | **Exam priority:** HIGH

```python
# Boilerplate code, easy to generate with any decent LLM

from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt

# Define label order explicitly (important for consistent output)
labels = ["CH", "FR", "DE"]

# ===== Example 1 =====
y_true_1 = ["CH", "FR", "DE", "CH", "FR"]
y_pred_1 = ["CH", "DE", "CH", "FR", "FR"]

print("=== Example 1 ===")
print("Classification Report:")
print(classification_report(y_true_1, y_pred_1, labels=labels, zero_division=0))

cm1 = confusion_matrix(y_true_1, y_pred_1, labels=labels)

disp1 = ConfusionMatrixDisplay(confusion_matrix=cm1, display_labels=labels)
disp1.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix – Example 1")
plt.show()



# ===== Example 2 =====
y_true_2 = ["CH", "FR", "DE", "CH", "FR"]
y_pred_2 = ["CH", "CH", "CH", "FR", "CH"]

print("\n=== Example 2 ===")
print("Classification Report:")
print(classification_report(y_true_2, y_pred_2, labels=labels, zero_division=0))

cm2 = confusion_matrix(y_true_2, y_pred_2, labels=labels)

disp2 = ConfusionMatrixDisplay(confusion_matrix=cm2, display_labels=labels)
disp2.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix – Example 2")
plt.show()
```
