# AdvNLP Lab 6: LLM Fine-Tuning with PEFT for emotion classification

**Objectives:** We train a classifier using the GPT-2 (small) model to classify text according to specific emotions such as joy, sadness, anger, and more.

To reduce the model size, we make use of LoRA (Low-Rank Adaptation) through the PEFT library.

**Infrastructure:** Training can be done on CPU, but it will be significantly faster on GPU. We recommend using the free GPU provided by Google Colab, or your own GPU if available.


**Useful documentation:** Since you'll use PEFT for this assignment, the respective [Hugging Face Quicktour](https://huggingface.co/docs/peft/quicktour) might be helpful.
Moreover, consider the Hugging Face tutorial on [text classification fine-tuning](https://huggingface.co/docs/transformers/en/tasks/sequence_classification).
